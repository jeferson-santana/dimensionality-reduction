```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
# carregar o dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
# scalonar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# aplicar pca para redução de dimensinalidade
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Visualizar o PCA
plt.figure(figsize=(8, 6))
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=target_name)
plt.title("PCA do Iris Dataset")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend()
plt.show()
```

```python
perplexity = 40
learning_rate = 1000
n_iter = 2000
```

```python
# aplicar o t-sne para redução de dimensionalidade
tsne = TSNE(n_components=2, random_state=42, perplexity=perplexity, learning_rate=learning_rate, n_iter=n_iter)
X_tsne = tsne.fit_transform(X_scaled)
```

```python
# Visualizar o t-SNE ajustado
plt.figure(figsize=(8, 6))
for i, target_name in enumerate(target_names):
    plt.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], label=target_name)
plt.title(f"t-SNE do Iris Dataset (Perplexity={perplexity}, LR={learning_rate}, Iter={n_iter})")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.legend()
plt.show()
```
